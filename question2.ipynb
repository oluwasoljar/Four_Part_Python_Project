```python
import wmi
import psutil
```

```python
def cpu_usage():  
    print('The CPU usage is: ', psutil.cpu_percent(10) , '%')
```

```python
# Initializing the wmi constructor
f = wmi.WMI()

def processes():
    process_list = []
    # Iterating through all the running processes and appending total to the list
    for i in range(len(f.Win32_Process())):
        process_list.append(i)

    return len(process_list)

cumm_list = []
cpu_usage = []

while True:
    try:
        print("====================================================================================")
        print("\nTO TERMINATE THE PROGRAM AT ANY POINT, PRESS THE 'STOP' BUTTON ONCE AND WAIT.\n")
        
        print("Checking C.P.U. usage, please wait...\n")
        cpu_pct = psutil.cpu_percent(20)
        print('The CPU usage is : ', cpu_pct , '%\n')
        
        cpu_usage.append(cpu_pct)
        print('List of total C.P.U. usage per iteration: ', cpu_usage)
        for x in range(len(cpu_usage)):
            dif = cpu_usage[x] - cpu_usage[x-1]
    
        if dif < 0:
            print(abs(dif), '% less C.P.U. usage now compared to previously\n')
        elif dif >= 10:
            print(dif, '% more C.P.U. usage now compared to previously\n')
            listOfProcessNames = list()
            # Iterate over all running processes
            for proc in psutil.process_iter():
                # Get process detail as dictionary
                pInfoDict = proc.as_dict(attrs=['name', 'cpu_percent'])
                
                # Append dict of process detail in list
                listOfProcessNames.append(pInfoDict)
                
                # Pick out processes with CPU usage equal to or more than 1%
                values = [d for d in listOfProcessNames if d['cpu_percent'] >= 1.0]
                existing_dicts = set()
                filtered_list = []
                for d in values:
                    if (d['name'], d['cpu_percent']) not in existing_dicts:
                        existing_dicts.add((d['name'], d['cpu_percent']))
                        filtered_list.append(d)
            print('The list of processes with usage at or over 1% are: ')
            print(*filtered_list, sep = "\n")
        else:
            print(dif, '% more C.P.U. usage now compared to previously\n')   
        x+=1            
        
        
        print("\nCalculating total number of processes running on P.C., please wait...\n")
        tot_process = processes()
        print('Number of processes running on P.C. is : ' , tot_process, ' processes\n')
        
        cumm_list.append(tot_process)
        print('List of total processes running on P.C. per iteration: ', cumm_list)
        for i in range(len(cumm_list)):
            diff = cumm_list[i] - cumm_list[i-1]
        
        if diff < 0:
            print(abs(diff), ' less processes running on computer now compared to previously\n')
        elif diff >= 0:
            print(diff, ' more processes running on computer now compared to previously\n')
        i+=1
    except KeyboardInterrupt:
        print("Process terminated")
        break
    


```
